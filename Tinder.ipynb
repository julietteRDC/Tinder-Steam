```python
print("Vérification des outliers pour les notes (1-10)")

# Colonnes à vérifier
rating_cols = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like', 'prob']

# Vérifier chaque colonne
for col in rating_cols:
    if col in df.columns:
        # Compter les valeurs < 1 et > 10
        inferieur_1 = (df[col] < 1).sum()
        supérieur_10 = (df[col] > 10).sum()
        total_outliers = inferieur_1 + supérieur_10

        print(f"{col}:")
        print(f"  Valeurs < 1: {inferieur_1}")
        print(f"  Valeurs > 10: {supérieur_10}")
        print(f"  Total outliers: {total_outliers}")
```

```python
# Supprimer les lignes avec des notes < 1 ou > 10

cols_concernees = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like', 'prob']
df = df[~((df[cols_concernees] < 1) | (df[cols_concernees] > 10)).any(axis=1)]
```

```python
## Effet de la confiance en soi

# Probabilité estimée vs décision réelle
reciprocite = df.groupby('dec')['prob'].mean()

# Confiance vs succès
df['confiance_elevee'] = df['prob'] > 7
confiance_resultats = pd.crosstab(df['confiance_elevee'], df['dec'])
confiance_resultats.index = ['Confiance faible (≤7)', 'Confiance élevée (>7)']
confiance_resultats.columns = ['Refus', 'Acceptation']

print("Résultats selon le niveau de confiance :\n")
print(confiance_resultats)


# Calcul des taux de succès
print("\nTaux de succès :")

for niveau in confiance_resultats.index:
    total = confiance_resultats.loc[niveau].sum()
    acceptations = confiance_resultats.loc[niveau, 'Acceptation']
    taux = (acceptations / total) * 100

    print(f"{niveau}: {acceptations}/{total} = {taux:.1f}%")


# Calcul de la différence
confiance_faible = confiance_resultats.loc['Confiance faible (≤7)', 'Acceptation'] / confiance_resultats.loc['Confiance faible (≤7)'].sum() * 100
confiance_elevee = confiance_resultats.loc['Confiance élevée (>7)', 'Acceptation'] / confiance_resultats.loc['Confiance élevée (>7)'].sum() * 100
difference = confiance_elevee - confiance_faible

print(f"\nDifférence de succès: +{difference:.1f} points de pourcentage")
```

```python
## Visiualisation pour montrer l'impact de la confiance en soi

# Calcul des taux de succès
taux = []
for niveau in confiance_resultats.index:
    total = confiance_resultats.loc[niveau].sum()
    acceptations = confiance_resultats.loc[niveau, 'Acceptation']
    taux_pct = (acceptations / total) * 100
    taux.append(taux_pct)

difference = taux[1] - taux[0]

# Créer le graphique
plt.figure(figsize=(10, 7))

categories = ['Confiance faible\n(≤7)', 'Confiance élevée\n(>7)']
colors = ['lightcoral', 'lightgreen']

bars = plt.bar(categories, taux, color=colors, alpha=0.8,
               edgecolor=['darkred', 'darkgreen'], linewidth=2)

# Pourcentages sur les barres
for bar, taux_pct in zip(bars, taux):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
            f'{taux_pct:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

# Mettre en évidence la différence
plt.text(0.5, max(taux) * 0.75, f'Différence: +{difference:.1f} points de pourcentage',
         ha='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.8', facecolor='gold', alpha=0.9))

# Personnalisation
plt.ylabel('Taux de succès (%)', fontsize=12)
plt.title('Impact de la confiance en soi sur le succès en speed dating',
          fontsize=14, fontweight='bold', pad=20)
plt.ylim(0, max(taux) * 1.2)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Les gens confiants, qui estiment avoir plus de 70% de chances ont 25 points de pourcentage de succès en plus. La confiance en soi améliore donc significativement les chances de succès.

```python
## Impact de l'objectif 'goal'

# Liste des différents ojectifs
liste_objectifs = {
    1: "Soirée sympa",
    2: "Rencontrer de nouvelles personnes",
    3: "Obtenir un rendez-vous",
    4: "Relation sérieuse",
    5: "Dire que je l'ai fait",
    6: "Autre"
}

# Taux de succès par objectif
if 'goal' in df.columns:
    goal_succes = df.groupby('goal')['dec'].agg(['count', 'mean'])
    goal_succes.columns = ['Nb_cas', 'Taux_acceptation']
    goal_succes['Taux_pourcentage'] = (goal_succes['Taux_acceptation'] * 100).round(1)
    print("Taux d'acceptation selon l'objectif de participation :")
    for goal_num, row in goal_succes.iterrows():
        if pd.notna(goal_num) and goal_num in liste_objectifs and row['Nb_cas'] >= 10:
            print(f"{liste_objectifs[int(goal_num)]}: {row['Taux_pourcentage']:.1f}% ({int(row['Nb_cas'])} cas)")

print("\n")

# Comparaison objectifs sérieux vs informel
if 'goal' in df.columns:
    # Créer des catégories
    df['goal_type'] = df['goal'].map({
        1: 'Informel',
        2: 'Informel',
        3: 'Sérieux',
        4: 'Sérieux',
        5: 'Informel',
        6: 'Autre'
    })

    goal_type_succes = df.groupby('goal_type')['dec'].agg(['count', 'mean'])
    goal_type_succes.columns = ['Nb_cas', 'Taux_acceptation']
    goal_type_succes['Taux_pourcentage'] = (goal_type_succes['Taux_acceptation'] * 100).round(1)

    print("Comparaison par type d'intention :")
    for goal_type, row in goal_type_succes.iterrows():
        if pd.notna(goal_type) and row['Nb_cas'] >= 10:
            print(f"{goal_type}: {row['Taux_pourcentage']:.1f}% ({int(row['Nb_cas'])} cas)")

    # Calculer la différence
    if 'Sérieux' in goal_type_succes.index and 'Informel' in goal_type_succes.index:
        diff = goal_type_succes.loc['Sérieux', 'Taux_pourcentage'] - goal_type_succes.loc['Informel', 'Taux_pourcentage']
        print(f"\nDifférence Sérieux vs Informel: {diff:+.1f} points de pourcentage")
```

```python
## Visualisation pour mettre en évidence le taux d'accepetation selon l'objectif rechercher

# Ordre personnalisé pour l'affichage
ordre_affichage = [1, 2, 5, 3, 4, 6]

# Préparer les données dans l'ordre voulu
objectifs_ordonnes = []
taux_ordonnes = []
couleurs = []

for goal_id in ordre_affichage:
    if goal_id in goal_succes.index and goal_succes.loc[goal_id, 'Nb_cas'] >= 10:
        objectifs_ordonnes.append(liste_objectifs[goal_id])
        taux_ordonnes.append(goal_succes.loc[goal_id, 'Taux_pourcentage'])

        # Définir les couleurs selon les catégories
        if goal_id in [1, 2, 5]:
            couleurs.append('lightsalmon')
        elif goal_id in [3, 4]:
            couleurs.append('lightgreen')
        else:
            couleurs.append('lightgray')

# Créer le graphique
plt.figure(figsize=(12, 7))

bars = plt.bar(range(len(objectifs_ordonnes)), taux_ordonnes, color=couleurs, alpha=0.8,
               edgecolor='black', linewidth=0.5)

# Ajouter les pourcentages sur les barres
for i, (bar, taux) in enumerate(zip(bars, taux_ordonnes)):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'{taux:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=11)

# Personnalisation
plt.xlabel('Objectifs de participation', fontweight='bold', fontsize=12)
plt.ylabel('Taux de succès (%)', fontweight='bold', fontsize=12)
plt.title('Taux de succès par objectif de participation', fontsize=16, fontweight='bold', pad=20)
plt.xticks(range(len(objectifs_ordonnes)), objectifs_ordonnes, rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.ylim(0, max(taux_ordonnes) * 1.15)

# Créer la légende
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='lightsalmon', alpha=0.8, label='Informel'),
    Patch(facecolor='lightgreen', alpha=0.8, label='Sérieux'),
    Patch(facecolor='lightgray', alpha=0.8, label='Autre')
]
plt.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.show()
```

Hypothèses : les gens sérieux sont plus investis/attentifs pendant les 4 minutes et ils rassurent ceux qui cherchent aussi quelque chose de solide.

```python
## L'impact de l'odre de passage

# Nombre de dates par participant
dates_par_participant = df['iid'].value_counts().sort_values(ascending=False)

print("Statistiques:")
print(f"Minimum de dates: {dates_par_participant.min()}")
print(f"Maximum de dates: {dates_par_participant.max()}")
print(f"Moyenne de dates: {dates_par_participant.mean():.1f}")
print(f"Médiane de dates: {dates_par_participant.median():.1f}")

print(f"\nDistribution:")
distribution = dates_par_participant.value_counts().sort_index()
for nb_dates, nb_participants in distribution.items():
    print(f"{nb_participants} participants ont eu {nb_dates} dates")


# Évolution des critères pendant la soirée
df['premiers_date'] = df['order'] <= 5
premiers_vs_derniers = df.groupby(['premiers_date', 'dec']).size().unstack()

premiers_vs_derniers.index = ['Derniers Date', 'Premiers Date']
premiers_vs_derniers.columns = ['REFUS', 'ACCEPTATION']

premiers_vs_derniers = premiers_vs_derniers.loc[['Premiers Date', 'Derniers Date']]
premiers_vs_derniers = premiers_vs_derniers[['ACCEPTATION','REFUS']]

print("\n")
print(premiers_vs_derniers)

# Calcul des taux d'acceptation
premiers_acceptation = premiers_vs_derniers.loc['Premiers Date', 'ACCEPTATION']
premiers_refus = premiers_vs_derniers.loc['Premiers Date', 'REFUS']
derniers_acceptation = premiers_vs_derniers.loc['Derniers Date', 'ACCEPTATION']
derniers_refus = premiers_vs_derniers.loc['Derniers Date', 'REFUS']


#Premiers rendez-vous (1-5)
total_premiers = premiers_refus + premiers_acceptation
taux_premiers = (premiers_acceptation / total_premiers) * 100

print("\nPremiers rendez-vous (1-5) :")
print(f"Taux d'acceptation : {premiers_acceptation}/{total_premiers} = {taux_premiers:.1f}%\n")


# Rendez-vous suivants (6+)
total_derniers = derniers_refus + derniers_acceptation
taux_derniers = (derniers_acceptation / total_derniers) * 100

print("Rendez-vous suivants (6+) :")
print(f"Taux d'acceptation : {derniers_acceptation}/{total_derniers} = {taux_derniers:.1f}%\n")


# Comparaison
difference = taux_premiers - taux_derniers
print(f"Différence : {taux_premiers:.1f}% - {taux_derniers:.1f}% = {difference:+.1f}%")
```

```python
## Visualisation pour montrer l'impact de l'odre de passage

# Créer le graphique
plt.figure(figsize=(10, 6))

categories = ['Premiers rendez-vous\n(1-5)', 'Rendez-vous suivants\n(6+)']
taux = [taux_premiers, taux_derniers]
colors = ['lightgreen', 'lightblue']

bars = plt.bar(categories, taux, color=colors, alpha=0.8,
               edgecolor=['darkgreen', 'darkblue'], linewidth=2)

# Pourcentages sur les barres
details = [
    f'{taux_premiers:.1f}%',
    f'{taux_derniers:.1f}%'
]\

for bar, detail in zip(bars, details):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            detail, ha='center', va='bottom', fontweight='bold', fontsize=11)

# Ligne de différence
difference = taux_premiers - taux_derniers
plt.text(0.5, max(taux) * 0.8, f'Différence: {difference:+.1f} points de pourcentage',
         ha='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.8))

plt.ylabel('Taux d\'acceptation (%)')
plt.title('Impact de l\'ordre des rendez-vous sur le taux d\'acceptation',
          fontsize=14, fontweight='bold', pad=15)
plt.ylim(0, max(taux) * 1.2)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Cela signifie que les gens sont légèrement plus sélectifs au fil de la soirée. Il faut donc mieux passer en début de soirée, mais l'impact reste limité.

**Pourquoi les gens acceptent un deuxième rendez-vous ?**

Les analyses révèlent que l'acceptation d'un deuxième rendez-vous dépend de quatre facteurs principaux classés par ordre d'importance :
1. L'attractivité physique :
Le facteur dominant reste l'attraction immédiate. Un critère universel mais plus déterminant chez les hommes que chez les femmes.
2. Le "fun factor" :
Être perçu comme amusant et divertissant est crucial. Les participants notés 7/10 ou plus en "fun" dépassent 50% de chances d'acceptation, prouvant que l'humour et la légèreté comptent énormément en 4 minutes.
3. Les intérêts partagés perçus :
Paradoxalement, c'est la perception d'avoir des points communs qui compte, pas la réalité mesurée. Il faut donc savoir créer une impression de connexion rapide.
4. L'état d'esprit et l'expérience :
- La confiance en soi double presque les chances de succès (64% vs 38%)
- L'expérience en dating améliore les performances (+13 points entre experts et novices)
- Les intentions sérieuses paradoxalement augmentent le succès (+4 points vs intentions casual)


Et enfin, pour mettre toutes les chances de son côté, mieux vaut passer en début de soirée : les premiers rendez-vous obtiennent +2.4 points d'acceptation grâce à l'effet de fraîcheur avant que la fatigue et la sélectivité s'installent.



Conclusion : En speed dating, on accepte quelqu'un qui nous attire, nous amuse, nous donne l'impression de compatibilité, et qui dégage une assurance authentique. L'art de réussir réside dans la capacité à créer rapidement cette alchimie en 4 minutes !
