```python
## Effet de la confiance en soi

# Probabilit estime vs dcision relle
reciprocite = df.groupby('dec')['prob'].mean()

# Confiance vs succs
df['confiance_elevee'] = df['prob'] > 7
confiance_resultats = pd.crosstab(df['confiance_elevee'], df['dec'])
confiance_resultats.index = ['Confiance faible (7)', 'Confiance leve (>7)']
confiance_resultats.columns = ['Refus', 'Acceptation']

print("Rsultats selon le niveau de confiance :\n")
print(confiance_resultats)


# Calcul des taux de succs
print("\nTaux de succs :")

for niveau in confiance_resultats.index:
    total = confiance_resultats.loc[niveau].sum()
    acceptations = confiance_resultats.loc[niveau, 'Acceptation']
    taux = (acceptations / total) * 100

    print(f"{niveau}: {acceptations}/{total} = {taux:.1f}%")


# Calcul de la diffrence
confiance_faible = confiance_resultats.loc['Confiance faible (7)', 'Acceptation'] / confiance_resultats.loc['Confiance faible (7)'].sum() * 100
confiance_elevee = confiance_resultats.loc['Confiance leve (>7)', 'Acceptation'] / confiance_resultats.loc['Confiance leve (>7)'].sum() * 100
difference = confiance_elevee - confiance_faible

print(f"\nDiffrence de succs: +{difference:.1f} points de pourcentage")
```

```python
## Visiualisation pour montrer l'impact de la confiance en soi

# Calcul des taux de succs
taux = []
for niveau in confiance_resultats.index:
    total = confiance_resultats.loc[niveau].sum()
    acceptations = confiance_resultats.loc[niveau, 'Acceptation']
    taux_pct = (acceptations / total) * 100
    taux.append(taux_pct)

difference = taux[1] - taux[0]

# Crer le graphique
plt.figure(figsize=(10, 7))

categories = ['Confiance faible\n(7)', 'Confiance leve\n(>7)']
colors = ['lightcoral', 'lightgreen']

bars = plt.bar(categories, taux, color=colors, alpha=0.8,
               edgecolor=['darkred', 'darkgreen'], linewidth=2)

# Pourcentages sur les barres
for bar, taux_pct in zip(bars, taux):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
            f'{taux_pct:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

# Mettre en vidence la diffrence
plt.text(0.5, max(taux) * 0.75, f'Diffrence: +{difference:.1f} points de pourcentage',
         ha='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.8', facecolor='gold', alpha=0.9))

# Personnalisation
plt.ylabel('Taux de succs (%)', fontsize=12)
plt.title('Impact de la confiance en soi sur le succs en speed dating',
          fontsize=14, fontweight='bold', pad=20)
plt.ylim(0, max(taux) * 1.2)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Les gens confiants, qui estiment avoir plus de 70% de chances ont 25 points de pourcentage de succs en plus. La confiance en soi amliore donc significativement les chances de succs.

```python
## L'impact de l'odre de passage

# Nombre de dates par participant
dates_par_participant = df['iid'].value_counts().sort_values(ascending=False)

print("Statistiques:")
print(f"Minimum de dates: {dates_par_participant.min()}")
print(f"Maximum de dates: {dates_par_participant.max()}")
print(f"Moyenne de dates: {dates_par_participant.mean():.1f}")
print(f"Mdiane de dates: {dates_par_participant.median():.1f}")

print(f"\nDistribution:")
distribution = dates_par_participant.value_counts().sort_index()
for nb_dates, nb_participants in distribution.items():
    print(f"{nb_participants} participants ont eu {nb_dates} dates")


# volution des critres pendant la soire
df['premiers_date'] = df['order'] <= 5
premiers_vs_derniers = df.groupby(['premiers_date', 'dec']).size().unstack()

premiers_vs_derniers.index = ['Derniers Date', 'Premiers Date']
premiers_vs_derniers.columns = ['REFUS', 'ACCEPTATION']

premiers_vs_derniers = premiers_vs_derniers.loc[['Premiers Date', 'Derniers Date']]
premiers_vs_derniers = premiers_vs_derniers[['ACCEPTATION','REFUS']]

print("\n")
print(premiers_vs_derniers)

# Calcul des taux d'acceptation
premiers_acceptation = premiers_vs_derniers.loc['Premiers Date', 'ACCEPTATION']
premiers_refus = premiers_vs_derniers.loc['Premiers Date', 'REFUS']
derniers_acceptation = premiers_vs_derniers.loc['Derniers Date', 'ACCEPTATION']
derniers_refus = premiers_vs_derniers.loc['Derniers Date', 'REFUS']


#Premiers rendez-vous (1-5)
total_premiers = premiers_refus + premiers_acceptation
taux_premiers = (premiers_acceptation / total_premiers) * 100

print("\nPremiers rendez-vous (1-5) :")
print(f"Taux d'acceptation : {premiers_acceptation}/{total_premiers} = {taux_premiers:.1f}%\n")


# Rendez-vous suivants (6+)
total_derniers = derniers_refus + derniers_acceptation
taux_derniers = (derniers_acceptation / total_derniers) * 100

print("Rendez-vous suivants (6+) :")
print(f"Taux d'acceptation : {derniers_acceptation}/{total_derniers} = {taux_derniers:.1f}%\n")


# Comparaison
difference = taux_premiers - taux_derniers
print(f"Diffrence : {taux_premiers:.1f}% - {taux_derniers:.1f}% = {difference:+.1f}%")
```

```python
## Visualisation pour montrer l'impact de l'odre de passage

# Crer le graphique
plt.figure(figsize=(10, 6))

categories = ['Premiers rendez-vous\n(1-5)', 'Rendez-vous suivants\n(6+)']
taux = [taux_premiers, taux_derniers]
colors = ['lightgreen', 'lightblue']

bars = plt.bar(categories, taux, color=colors, alpha=0.8,
               edgecolor=['darkgreen', 'darkblue'], linewidth=2)

# Pourcentages sur les barres
details = [
    f'{taux_premiers:.1f}%',
    f'{taux_derniers:.1f}%'
]\

for bar, detail in zip(bars, details):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            detail, ha='center', va='bottom', fontweight='bold', fontsize=11)

# Ligne de diffrence
difference = taux_premiers - taux_derniers
plt.text(0.5, max(taux) * 0.8, f'Diffrence: {difference:+.1f} points de pourcentage',
         ha='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.8))

plt.ylabel('Taux d\'acceptation (%)')
plt.title('Impact de l\'ordre des rendez-vous sur le taux d\'acceptation',
          fontsize=14, fontweight='bold', pad=15)
plt.ylim(0, max(taux) * 1.2)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Cela signifie que les gens sont lgrement plus slectifs au fil de la soire. Il faut donc mieux passer en dbut de soire, mais l'impact reste limit.
