# Analyse au niveau « macro »

Quel éditeur a publié le plus de jeux sur Steam ?

```python
top_publishers = df.groupBy("publisher").count().orderBy(desc("count"))
top_publishers.limit(1).toPandas()
```

Quels sont les jeux les mieux notés ?

```python
df.groupBy("name") \
  .agg(F.sum("positive").alias("total_positive")) \
    .orderBy(F.desc("total_positive")) \
      .limit(5).toPandas()
```

Y a-t-il eu plus de sorties de jeux pendant certaines années ?

```python
df = df.withColumn("release_date", to_date("release_date", "yyyy/MM/d"))

df = df.withColumn("year", year("release_date"))

df.groupBy("year") \
  .agg(F.count("year").alias("total_year")) \
    .orderBy(F.desc("total_year")) \
      .limit(5).toPandas()
```

Comment sont distribués les prix ? Y a-t-il beaucoup de jeux avec une réduction ?

```python
df.select("price", "discount").printSchema()
```

```python
df = df.withColumn("price", col("price").cast("float")) \
       .withColumn("discount", col("discount").cast("float"))
```

```python
df.select("price", "discount").printSchema()
```

```python
display(df.select('price').describe())
```

```python
filtered_price = df.filter(col("price") < 2500)
display(filtered_price.select("price"))
```

```python
filtered_discount = df.filter(col("discount") > 0)
nb_reduc = filtered_discount.select('discount').count()
rate_reduc = 100* nb_reduc / df.count()
print("Nombre de jeux en réduction : ", nb_reduc)
print(f"Pourcentage de jeux en réduction : {rate_reduc:.2f} %")
```

Quelles sont les langues les plus représentées ?

```python
split_df = df.withColumn("languages", split(F.col("languages"), ", "))
exploded_df = split_df.withColumn("language", explode(F.col("languages")))

display(exploded_df.groupBy("language").count().orderBy(F.desc('count')))
```


Y a-t-il beaucoup de jeux interdits aux moins de 16/18 ans ?

```python
df.select("required_age").printSchema()
```

```python
df = df.withColumn("required_age", col("required_age").cast("int"))
```

```python
df.select("required_age").printSchema()
```

```python
# Jeux pour les plus de 16 ans
game_more_16 = df.filter(col("required_age") >= 16)
nb_game_more_16 = game_more_16.count()

# Jeux pour les plus de 18 ans
game_more_18 = df.filter(col("required_age") >= 18)
nb_game_more_18 = game_more_18.count()

# Nombre total de jeux
nb_total = df.count()

# Pourcentages
rate_more_16 = (nb_game_more_16 / nb_total) * 100
rate_more_18 = (nb_game_more_18 / nb_total) * 100

print(f"Nombre de jeux interdits aux moins de 16 ans : {nb_game_more_16}, soit {rate_more_16:.2f}%")
print(f"Nombre de jeux interdits aux moins de 18 ans : {nb_game_more_18}, soit {rate_more_18:.2f}%")
```

# Analyse de la plateforme
# 

La plupart des jeux sont-ils disponibles sur Windows/Mac/Linux ?

```python
df.select(
  F.sum(col("windows").cast("int")).alias("Windows"),
  F.sum(col("mac").cast("int")).alias("Mac"),
  F.sum(col("linux").cast("int")).alias("Linux")).show()
```

```python
total = df.count()
rate_platform = df.select(
  F.sum(col("windows").cast("int")/ total * 100).alias("Windows"),
  F.sum(col("mac").cast("int")/ total * 100).alias("Mac"),
  F.sum(col("linux").cast("int")/ total * 100).alias("Linux"))
        
transposed_rate_platform = rate_platform.selectExpr("stack(3, 'Windows', Windows, 'Mac', Mac, 'Linux', Linux) as (Platform, Percentage)")

display(transposed_rate_platform)
```

Certains genres ont-ils tendance à être disponibles préférentiellement sur certaines plateformes ?

```python
df_genre.groupBy("genre").agg(
    avg(col("windows").cast("int")).alias("dispo_windows"),
    avg(col("mac").cast("int")).alias("dispo_mac"),
    avg(col("linux").cast("int")).alias("dispo_linux")
).orderBy(desc("dispo_windows")).show()
```

```python
# Disponibilités moyennes par genre
df_dispo = df_genre.groupBy("genre").agg(
    avg(col("windows").cast("int")).alias("dispo_windows"),
    avg(col("mac").cast("int")).alias("dispo_mac"),
    avg(col("linux").cast("int")).alias("dispo_linux")
)

# Genre le plus disponible
top_windows = df_dispo.orderBy(desc("dispo_windows")).limit(1)
top_mac = df_dispo.orderBy(desc("dispo_mac")).limit(1)
top_linux = df_dispo.orderBy(desc("dispo_linux")).limit(1)

print("Top genre Windows:")
top_windows.show()

print("Top genre Mac:")
top_mac.show()

print("Top genre Linux:")
top_linux.show()
```
